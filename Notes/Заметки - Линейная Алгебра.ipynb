## Определитель

### Формула

Определетиль - сумма произведений элементов по одному из каждого столбца и каждой строки по всем возможным перестановкам, взятых со знаком четности перестановки

$\det A = \sum_{\alpha} (-1)^{\sigma(\alpha)} a_{1\alpha_1} a_{2\alpha_2} \cdots a_{n\alpha_n}$

### Свойства

$|E| = 1$

$|A| = |A^T|$

Если есть одинаковые строки, то $|A| = 0$

Если есть нулевые строки, то $|A| = 0$

Смена строк ведет к изменению знака определетеля. Это связано с изменением четности инверсий

#### Линейность определителя

$
\begin{vmatrix}
\cdot & k \cdot a_1 & \cdot \\
\cdot & k \cdot a_2 & \cdot \\
\cdot & \cdot & \cdot \\
\end{vmatrix}
=
k \cdot
\begin{vmatrix}
\cdot & a_1 & \cdot \\
\cdot & a_2 & \cdot \\
\cdot & \cdot & \cdot \\
\end{vmatrix}
$

$
\begin{vmatrix}
\cdot & b_1 + c_1 & \cdot \\
\cdot & b_2 + c_2 & \cdot \\
\cdot & \cdot & \cdot \\
\end{vmatrix}
=
\begin{vmatrix}
\cdot & b_1 & \cdot \\
\cdot & b_2 & \cdot \\
\cdot & \cdot & \cdot \\
\end{vmatrix}
+
\begin{vmatrix}
\cdot & c_1 & \cdot \\
\cdot & c_2 & \cdot \\
\cdot & \cdot & \cdot \\
\end{vmatrix}
$

Из данных свойств следует, что добавление или вычитание из строки или столбца линейной комбинации других строк или столбцов не изменит определителя.

## Обратная матрица

Обратная матрица - это такая матрица, что $AA^-1 = E$

$A^{-1} = \hat{A} \dfrac{1}{\det A}$

Отсюда можно заключить, что обратная матрица существует только у невырожденных матриц. Иными словами, при применение вырожденного оператора часть информации теряется и единого обратного пути нет. 

На практике удобнее находить обратную матрицу методом Гаусса (элементарных преобразований). Как известно, составную матрицу $(A \mid E)$ преобразуют до вида $(E \mid A^{-1})$. Почему так, можно увидеть из следующих соображений. Метод Гаусса - это применение слева матриц элементарных преобразований $L$. Тогда

$ L_1 \cdot L_2 \cdots L_n \cdot A = E$

$ A^{-1} \cdot A = E$

$ L_1 \cdot L_2 \cdots L_n = A^{-1}$
