# Заметки по теории групп

## Группы

**Группа** - это непустое множество с заданной групповой операцией, удовлетворяющее аксимомам группы.

Аксиомы группы

- Групповая операция ассоциативна: $a*(b*c) = (a*b)*c$
- Существует нейтральный элемент $e$, такой что  $a*e = e*a = a$
- Для каждого элемента $a$ существует симметричный $a^{-1}$, такой что $a * a^{-1} = e$

Группа обладает свойством, что результат групповой операции всегда лежит в группе.

**Абелева группа** - группа, обладающая свойством коммутативности

**Подгруппа** - непустое множество группы $G$, которое само является группой относительно операции в $G$

**Теорема**. Подмножество $H$ группы $G$ является подгруппой, если выполняются следующие импликации:

1) $a, b \in H \Rightarrow ab \in H$

2) $a\in H \Rightarrow a^{-1} \in H$

**Смежный класс** - множество вида $aH$ или $Ha$, где $H$ - подгруппа группы $G$, а $a \in G$

**Конечная группа** - группа с конечным числом элементов

**Циклическая группа** - подгруппа группы $G$, порожденная целыми степенями элемента $a \in G$. Степень определяется как:

$a^n = \begin{cases}
aa\dots a,\:n > 0\\
1,\:n = 0\\
(a^{-1})^m ,\:m = -n,\:n < 0
\end{cases}$

## Морфизмы групп

Изоморфизмом групп $G_1$ и $G_2$ называется отображение $\:f:G_1 \rightarrow G_2$, обладающее следующими свойствами

1) Гомоморфизм (сохранение групповой операции): $f(a*_1b) = f(a) *_2 f(b)$

2) Отображение $f$ биективно

Обозначение $G_1 \cong G_2$

## Кольцо и поле

**Кольцо** - непустое множество $K$ с заданными двумя операциями ($+$ и $*$), обладающее следующими свойствами

1. $a + b = b + a$
2. $(a + b) + c = a + (b + c)$
3. $\exists 0 \in K: a + 0 = 0 + a = a$
4. $\forall a \in K, \: \exists -a \in K : a + (-a) = (-a) + a = 0$
5. $(ab)c = a(bc)$
6. $(a + b)c = ab + bc,\: a(b + c) = ab + ac$

**Коммутативное кольцо** - кольцо с коммутативным умножением

**Кольцо с единицей** - существует нейтральный элемент по умножению

**Поле** - коммутативное кольцо с единицей, в котором каждый отличный от нуля элемент имеет обратный элемент
