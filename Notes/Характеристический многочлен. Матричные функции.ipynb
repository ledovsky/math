# Характеристический многочлен. Матричные фукнции


## Характеристический многочлен

**Аннулирующий многочлен.** Такой скаларный многочлен $P(\lambda)$, что $P(A) = 0$

**Характеристический многочлен.** Задается, как $\det(A - \lambda E)$. Является аннулирующим (теор. Гамильтона-Кэли)

**Минимальный многочлен.** Аннулирующий многочлен минимальной степени с коэффициентом при старшей степени, равным 1. Является единственным. Может отличаться от характеристического многочлена

### Обобщенная теорема Безу

**Теорема.** Остаток от правого (левого) деления матричного многочлена $F(\lambda)$ на бином $(\lambda E - A)$ равен $F(A)$

**Логика доказательства**

$$F(\lambda) = Q(\lambda) \cdot (\lambda E - A) + R$$

$R$ не зависит от $\lambda$, т.к. при делении многочленов остаток имеет меньшую степень, чем делитель

Тогда $$F(A) = Q(A) \cdot (A - A) + R = R$$

### Теорема Гамильтона-Кэли

**Теорема**. Всякая квадратная матрица $A$ удовлетворяет своему характеристическому уравненю

$$\phi (A) = 0$$

**Логика доказательства**

Обозначим $B(\lambda)$ присоединенную матрицу (т.е. матрицу алгебраических дополнений) для $(\lambda E - A)$

Из определения обратной матрицы

$$\frac{1}{(\lambda E - A)} = \frac{1}{\det{(\lambda E - A)}} \cdot B(\lambda)$$

$$\det{(\lambda E - A)} \cdot E = \phi(\lambda) = (\lambda E - A) \cdot B(\lambda)$$

Последнее выражение означает, что $\phi(\lambda)$ без остатка делится на $(\lambda E - A)$

По Обобщенной Теореме Безу остаток от деления равен $\phi(А)$. Значит $\phi (A) = 0$ $\blacksquare$

### Полином от матрицы. Собственные значения

**Теорема.** Пусть $\lambda_1, \cdots, \lambda_n$ - собственные числа матрицы $A$, а $g(u)$ - некоторый скалярный многочлен. Тогда собственные числа матрицы $g(A)$ будут равны $g(\lambda_1), \cdots, g(\lambda_n)$

###  Нахождение присоединенной матрицы с помощью характеристического многочлена

**Формула.** Пусть $\phi(\lambda) = \det(A - \lambda E) = \lambda^n + p_1 \cdot \lambda^{n-1} + p_2 \cdot \lambda^{n-2} + \dots$

Тогда $$ B(\lambda) = \delta(\lambda E, A)$$

$$ \delta(\lambda E, A) = \frac{\phi(\lambda E) - \phi(A)}{\lambda E - A} = \lambda^{n-1} + (A - p_1) \cdot \lambda^{n-1} + (A^2 - p_1 A - p_2) \cdot \lambda^{n-2} + \dots$$

**Логика доказательства.**

Функционал $\delta(\lambda, \mu)$ выводится напрямую. Необходимо аккуратно выразить разность $\phi(\lambda) - \phi(\mu)$ так, чтобы представить многочлен относительно $\lambda$

Далее, подставив в функционал $\lambda E$ и $A$, $\phi(A)$ обращается в 0, и получается формула из доказательства теоремы Гамильтона-Кэли

$$\det{(\lambda E - A)} \cdot E = (\lambda E - A) \cdot B(\lambda) = (\lambda E - A) \cdot \delta(\lambda E, A)$$

Откуда следует, что 

$$ B(\lambda) = \delta(\lambda E, A)$$

### Вычисление минимального многочлена

 
$$\psi(\lambda) = \frac{det(A - \lambda E)}{D_{n-1}}$$

где  $D_{n-1}$ - наибольший общий делитель элементов матрицы $B(\lambda)$ алгебраических дополнений $(A - \lambda E)$

**Логика доказательства.**

Доказательство довольно длинное, но невооруженным взглядом видно из доказательства теоремы Гамильтона-Кэли, что характеристический многочлен можно сократить на делитель $B(\lambda)$
